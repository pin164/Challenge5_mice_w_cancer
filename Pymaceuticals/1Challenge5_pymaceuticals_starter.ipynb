# 1. The datasets are merged into a single DataFrame. 

```python
# Combine the data into a single DataFrame 
# use left join to ensure all meta data has study results

mouse_merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="left")
mouse_merge_outer_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview

print("left join mouse results:\n", mouse_merge_df)
print("outer join mouse results:\n", mouse_merge_outer_df)

```

# 2.The number of mice are shown from the merged DataFrame

```python
# Checking the number of mice.  This is ambiguous.

mice_datapoint_count = len(mouse_merge_df)
mice_studied = mouse_merge_df["Mouse ID"].nunique()
print(f"number of mouse datapoints collected: {mice_datapoint_count}")
print(f"number of mice identified: {mice_studied}")
```

# 3.Each duplicate mice is found based on the Mouse ID and Timepoint.

```python
# Set the index of the DataFrame to be the combination of 'Mouse ID' and 'Timepoint'
mouse_merge_df.set_index(['Mouse ID', 'Timepoint'], inplace=False)
print(f"Index by ID & Timepoint:\n {mouse_merge_df}")

# Get the duplicate mouse entries by searching for duplicates in the index
# No need to specify subset since 'Mouse ID' and 'Timepoint' are now the index
duplicate_mice_data = mouse_merge_df[mouse_merge_df.index.duplicated(keep=False)]

# Get the unique Mouse IDs for which duplicate Timepoints exist using the index
duplicate_mouse_ids = duplicate_mice_data.index.get_level_values('Mouse ID').unique()
print(f"Duplicate mouse IDs:\n {duplicate_mouse_ids}")

# Display the duplicate mouse data entries
print(f"Duplicate mouse data entries:\n {duplicate_mice_data}")
```
